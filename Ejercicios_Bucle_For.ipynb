### 5.2. Calcular la suma de los primeros `n` números

#### Objetivo:
Sumar todos los números enteros desde 1 hasta `n` usando un bucle `for`.

#### Instrucciones:
1. Pide al usuario un número entero `n` con `input()`.
2. Usa un bucle `for` con `range()` para recorrer los números desde `1` hasta `n` (incluido).
3. Ve sumando cada número en una variable acumuladora.
4. Muestra el resultado final por pantalla.

```python
n = int(input("Introduce un numero entero:"))
suma_total = 0
for i in range(1,n+1):
    suma_total += i
    print(suma_total)

print("La suma de los numeros de 1 a",n, "es:",suma_total)
```

```python
# Ejemplo:
# Si el usuario introduce 5, se suman: 1 + 2 + 3 + 4 + 5 = 15

n = int(input("Introduce un número: "))
suma_total = 0

for i in range(1,n+1):
    suma_total += i #equivalente a suma_total = suma_total + i
    print(suma_total)

print("La suma de los números de 1 a", n, "es:", suma_total)
```

### 5.3. Contar vocales en una palabra

#### Objetivo:
Contar cuántas vocales contiene una palabra introducida por el usuario.

#### Instrucciones:
1. Pide al usuario una palabra con `input()`.
2. Recorre la palabra usando un bucle `for`.
3. Verifica si cada letra es una vocal (`a`, `e`, `i`, `o`, `u`).
4. Lleva la cuenta de cuántas vocales hay.
5. Muestra el resultado final.

```python
palabra = str(input("Introduce una palabra porfa"))
vocales = "aeiou"
contador = 0

for letras in palabra.lower()
    print("Letra actual:")
    if letras in vocales:  #este if es para contar las vocales,si la letra actual está en vocales
        contador += 1
        print()
```

```python
palabra = input("Introduce una palabra: ")
vocales = "aeiou"
contador = 0

for letter in palabra.lower():#metemos aqui el lower para que no importe si la letra es mayuscula o minuscula
    print(f"Letra: {letter}")
    if letter in vocales:
        contador += 1 # contador = contador + 1
        print(f"Vocal: {letter}, contador: {contador}")

print(f"Número de vocales en la palabra {palabra}: {contador}")
```

### 5.4. Multiplicando cada elemento de una lista por 2

#### Objetivo:
Crear una nueva lista que contenga el doble de cada número de la lista original.

#### Instrucciones:
1. Define una lista con varios números enteros.
2. Usa un bucle `for` para recorrer cada número.
3. Multiplica cada número por 2.
4. Guarda el resultado en una **nueva lista**.
5. Muestra ambas listas (la original y la nueva).

```python
lista_numeros = [4, 7, 11, 16, 23]
multiplicador = []

for n in lista_numeros:
    print(f"Numero actual,{n}")

n*2 for n in lista_numeros: #esto traducido seria "para cada n en lista_numeros, multiplica n por 2"
    multiplicador.append(n*2)
    print(f"El numero {n} multiplicado por 2 es: {multiplicador[-1]}")


```

```python
lista_numeros = [2, 13, 29, 8]
resultado = [] #se ponen listas vacias cuando queremos crear una lista nueva a partir de otra

for n in lista_numeros:
    resultado.append(n*2)
    print(resultado)

# equivalente a resultado = [n*2 for n in lista_numeros]
print("Lista original:", lista_numeros)
print("Lista multiplicada por 2:", resultado)
```
