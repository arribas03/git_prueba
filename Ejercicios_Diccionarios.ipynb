### 1. Diccionario de traducciones

Imagina que estás creando un pequeño **diccionario bilingüe** (Español → Inglés).  
1. Crea un diccionario llamado `traducciones` con **3 palabras** como claves en español (gato, perro, mesa) y sus traducciones en inglés (cat, dog, table) como valores.  
2. Muestra todas las **claves** (palabras en español) al usuario y pídele que introduzca una palabra en español (con `input()`).  
3. Comprueba si la palabra está en el diccionario:  
   - Si **existe**, muestra su traducción en inglés.  
   - Si **no existe**, indica que no se encuentra en el diccionario.  
4. Pregunta al usuario si desea **agregar** esa palabra al diccionario. Si responde que sí, pídele la traducción y añádela al diccionario.  
5. Muestra el diccionario final.

*Pistas:*
- Usa `in` para verificar si una clave existe en el diccionario.  
- Recuerda que para **insertar** o **actualizar** un par clave-valor se puede hacer con `diccionario[clave] = valor`.


````python
traducciones = {"gato": "cat", "perro": "dog", "mesa": "table"}
print("Palabras disponibles:", list(traducciones.keys())) #keys() devuelve las claves del diccionario
palabra = input("Introduce una palabra en español: ").lower()
if palabra in traducciones:
    print(f"La traducción de '{palabra}' es '{traducciones[palabra]}'")
else:
    print(f"La palabra '{palabra}' no está en el diccionario de traducciones.")
    nueva_palbra = input("¿Quieres añadirla? (s/n): ").lower()
    if nueva_palbra == 's':
        traduccion = input(f"Introduce la traducción de '{palabra}': ")
        traducciones[palabra] = traduccion
        print(f"Palabra '{palabra}' añadida con la traducción '{traduccion}'.")
        print("Diccionario actualizado:", traducciones)
    










````

````python
traducciones = {"gato": "cat", "perro": "dog", "mesa": "table"}

print("Palabras disponibles:", list(traducciones.keys())) #keys() devuelve las claves del diccionario
palabra = input("Introduce una palabra en español: ").lower()

if palabra in traducciones:
    print(f"La traducción es: {traducciones[palabra]}")
else:
    print("No se encuentra en el diccionario.")
    agregar = input("¿Quieres agregarla? (s/n): ")
    if agregar.lower() == "s":
        traduccion = input(f"Dime la traducción de {palabra}: ")
        traducciones[palabra] = traduccion
        print("Diccionario actualizado:", traducciones)
````

### 2. Registro de notas de alumnos

Quieres llevar un pequeño **registro** de los alumnos y sus **notas**.  
1. Crea un diccionario vacío llamado `notas_alumnos`.  
2. En un **bucle** `while True`, pide al usuario que introduzca:  
   - El nombre del alumno.  
   - La nota del alumno (un número).  
3. Almacena esa información en el diccionario `notas_alumnos` (la clave será el nombre y el valor será la nota).  
4. Pregunta al usuario si desea **agregar otro alumno**.  
   - Si contesta `"no"` (o `"No"`, ignorando mayúsculas/minúsculas), rompe el bucle.  
5. Después de salir del bucle, recorre el diccionario y **muestra** por pantalla un mensaje del tipo:  

````python
notas_alumnos = {}
while True:
    nombre_alumno = input("Introduce el nombre del alumno (o 'salir' para terminar): ")
    nota_alumno = float(input("Introduce la nota del alumno: ")) #float primera convierte la entrada en un número decimal
    notas_alumnos[nombre_alumno] = nota_alumno #la clave es el nombre del alumno y el valor es la nota
    agregar = input("¿Quieres añadir otro alumno? (s/n): ")
    if agregar.lower() == "n":
        break
for nombre,nota in notas_alumnos.items(): #items() devuelve una lista de tuplas (clave, valor)
    print(f"Alumno: {nombre} - Nota: {nota}")
    
````

````python
notas_alumnos = {}

while True:
    nombre = input("Nombre del alumno: ")
    nota = float(input("Nota del alumno: ")) # Convertimos a float
    notas_alumnos[nombre] = nota
    
    continuar = input("¿Agregar otro? (si/no): ")
    if continuar.lower() == "no":
        break

print("\n--- REGISTRO FINAL ---")
for nombre, nota in notas_alumnos.items():
    print(f"Alumno: {nombre} - Nota: {nota}")
````

### 3. Estadísticas de caracteres en una cadena

1. Pide al usuario que escriba una **frase** (o varias).  
2. Crea un diccionario llamado `conteo_caracteres` que guarde la **cantidad de veces** que aparece cada carácter en la frase.  
- Para cada carácter en la frase:  
  - Si ya existe como clave, incrementa su valor.  
  - Si no, añádelo con valor `1`.  
3. Ignora mayúsculas/minúsculas convirtiendo la frase a minúsculas antes de contar.  
4. Muestra el resultado en pantalla, por ejemplo:  

````python
frases = input("Introduce varias frases separadas por comas: ").lower().replace(" ","")
conteo_caracteres = {} 
for caracter in frases:
    if caracter in conteo_caracteres:
        conteo_caracteres[caracter] += 1 #es lo mismo que conteo_caracteres[caracter] = conteo_caracteres[caracter] + 1
    else:
        conteo_caracteres[caracter] = 1
    print(f"Carácter '{caracter}' aparece {conteo_caracteres[caracter]} veces")
print("Conteo final de caracteres:", conteo_caracteres) #para ordenar por cantidad de apariciones se


````

````python
frase = input("Escribe una frase: ").lower()
conteo = {}

for caracter in frase:
    if caracter == " ": # Opcional: ignorar espacios
        continue
        
    if caracter in conteo:
        conteo[caracter] += 1
    else:
        conteo[caracter] = 1

print("Frecuencia de caracteres:", conteo)
````

### 4. Control de inventario

Imagina que gestionas un almacén con productos y cantidades en stock.  
1. Define un diccionario `inventario` con algunas **claves** como `"manzanas"`, `"naranjas"`, `"leche"`, asociadas a números que representen la **cantidad** disponible.  
2. Muestra un **menú** (usando un bucle `while`) con opciones:  
1. `"1. Ver inventario"`  
2. `"2. Añadir producto"`  
3. `"3. Actualizar stock de un producto"`  
4. `"4. Eliminar producto"`  
5. `"5. Salir"`  
3. Según la opción elegida:  
- **Ver inventario**: Recorre el diccionario e imprime producto y cantidad.  
- **Añadir producto**: Pide el nombre del producto y la cantidad. Añádelo al diccionario (si ya existía, avisa al usuario o actualiza la cantidad).  
- **Actualizar stock**: Pide el nombre del producto. Si existe, pide la nueva cantidad y actualiza el valor en el diccionario. Si no existe, indícalo por pantalla.  
- **Eliminar producto**: Pide el nombre del producto y, si existe en el diccionario, elimínalo. Si no existe, indícalo.  
- **Salir**: Rompe el bucle.  
4. Muestra el diccionario final cuando el usuario salga.

````python
inventario = {"manzanas": 50, "naranjas": 30, "plátanos": 22}
while True:
    print("\n--- Gestión de Inventario ---\n") #esto se escribe asi para hacer un salto de línea, es decir, dejar una línea en blanco
    print("1. Mostrar inventario")
    print("2. Añadir producto")
    print("3. Actualizar stock")
    print("4. Eliminar producto")
    print("5. Salir")
    opcion = input("Selecciona una opción (1-5): ")
    if opcion == "1":
        print("Inventario actual:", inventario)
    elif opcion == "2":
        producto = input("Nombre del producto a añadir: ")
        cantidad = int(input("Cantidad a añadir: "))
        if producto in inventario:
            inventario[producto] += cantidad
            print(f"El producto '{producto}' ya existe. Actualizando cantidad.")
        else:
            inventario[producto] = cantidad
            print(f"Producto '{producto}' añadido con cantidad {cantidad}.")
    elif opcion == "3":
        producto = input("Nombre del producto a actualizar: ")
        if producto in inventario:
            cantidad = int(input("Nueva cantidad: "))
            inventario[producto] = cantidad
            print(f"Producto '{producto}' actualizado a cantidad {cantidad}.")
        else:
            print(f"El producto '{producto}' no existe en el inventario.")
    elif opcion == "4":
        producto = input("Nombre del producto a eliminar: ")
        if producto in inventario:
            del inventario[producto]
            print(f"Producto '{producto}' eliminado del inventario.")
        else:
            print(f"El producto '{producto}' no existe en el inventario.")
    elif opcion == "5":
        print("Saliendo del programa de gestión de inventario.")
        break
    print("Inventario actualizado:", inventario)
````

````python
inventario = {"manzanas": 50, "naranjas": 30, "leche": 20}

while True:
    print("\n--- GESTIÓN DE ALMACÉN ---")
    print("1. Ver inventario")
    print("2. Añadir producto")
    print("3. Actualizar stock")
    print("4. Eliminar producto")
    print("5. Salir")
    
    opcion = input("Elige una opción: ")
    
    if opcion == "1":
        for prod, cant in inventario.items():
            print(f"- {prod}: {cant}")
            
    elif opcion == "2":
        nombre = input("Nombre del producto: ")
        cantidad = int(input("Cantidad: "))
        inventario[nombre] = cantidad # Si existe lo sobrescribe, si no lo crea
        print("Producto guardado.")
        
    elif opcion == "3":
        nombre = input("Nombre del producto a actualizar: ")
        if nombre in inventario:
            cantidad = int(input("Nueva cantidad: "))
            inventario[nombre] = cantidad
        else:
            print("Error: El producto no existe.")
            
    elif opcion == "4":
        nombre = input("Nombre a eliminar: ")
        if nombre in inventario:
            del inventario[nombre] # O usar inventario.pop(nombre)
            print("Eliminado.")
        else:
            print("No encontrado.")
            
    elif opcion == "5":
        print("Saliendo...")
        break
    else:
        print("Opción no válida")
````

Imagina que tienes un diccionario que relaciona nombres de productos con sus precios, por ejemplo:

```python
precios = {
    "manzanas": 1.5,
    "naranjas": 2.0,
    "leche": 2.5
}
```
El usuario te pide saber el precio de "naranjas". Muéstralo por pantalla (usando get o []).

Añade un nuevo producto "yogur" con un precio de 1.2.

Cambia el precio de las "manzanas" a 1.75.

Elimina el producto "leche" de la lista porque ya no lo vas a vender.

Muestra todos los productos disponibles (claves) y sus precios (valores) de manera clara.


````python
precios = {"manzanas": 1.5, "naranjas": 2.0, "leche": 2.5}
precios.get("naranjas")
print(f"El precio de las naranjas es de {precios.get('naranjas')} euros el kilo")
nuevo_producto = "yogur"
nuevo_precio = 1.2
precios[nuevo_producto] = nuevo_precio # es lo mismo que precios.update({nuevo_producto: nuevo_precio})
precios.update({"manzanas": 1.75})  # Otra forma de añadir o actualizar
precios.pop("leche")  # Eliminar un producto
print("Precios actualizados:", precios)
````

````python
precios = {"manzanas": 1.5, "naranjas": 2.0, "leche": 2.5}

print("Precio naranjas:", precios.get("naranjas")) # O precios["naranjas"]

precios["yogur"] = 1.2
precios["manzanas"] = 1.75
if "leche" in precios:
    del precios["leche"]

print("Lista final:")
for prod, precio in precios.items():
    print(f"{prod}: {precio}€")
````
