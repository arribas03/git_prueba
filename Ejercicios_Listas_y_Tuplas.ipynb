### 1. Gestión de inventario de productos

1. Crea una lista llamada `inventario` que contenga algunos nombres de productos, por ejemplo:  
   
   `inventario = ["manzanas", "naranjas", "peras", "leche"]`

Decide que necesitas añadir "huevos" al final de la lista. Usa el método append().

Realiza un slicing para obtener los productos que van desde el índice 1 al 3 (sin incluir el 3) e imprímelos.

Resulta que te has quedado sin "leche" y debes eliminarla de la lista. Usa remove() para quitar "leche".

Te han enviado otra lista llamada nuevos_productos = ["café", "azúcar"]. Usa extend() para añadirlos al inventario.

Finalmente, muestra la lista resultante y su longitud.


````python
iventario = ["manzanas", "naranjas", "plátanos", "peras"]
iventario.append("Huevos")  # Agrega "Huevos" al final de la lista
print(iventario[1:3])
iventario.remove("naranjas")
iventario.extend(["cafe", "azucar"])
print(iventario)
print(len(iventario))  
````

````python
# 1. Crear la lista inicial
inventario = ["manzanas", "naranjas", "peras", "leche"]

# 2. Añadir "huevos" al final
inventario.append("huevos")
print("Inventario tras append:", inventario)

# 3. Slicing del índice 1 al 3 (sin incluir el 3)
# Recuerda: índice 1 es "naranjas", índice 2 es "peras".
sub_lista = inventario[1:3]
print("Productos seleccionados (1 al 3):", sub_lista)

# 4. Eliminar "leche"
if "leche" in inventario: # Es buena práctica comprobar antes
    inventario.remove("leche")
print("Inventario tras eliminar leche:", inventario)

# 5. Añadir nuevos productos con extend
nuevos_productos = ["café", "azúcar"]
inventario.extend(nuevos_productos)

# 6. Mostrar resultado final y longitud
print("Inventario final:", inventario)
print("Total de productos:", len(inventario))
````

### 2. Calcular estadísticas de notas
Imagina que estás gestionando las notas de un grupo de estudiantes.

Crea una lista llamada notas con valores numéricos que representen distintas calificaciones, por ejemplo:

`notas = [7, 9, 5, 10, 6, 7, 8, 9]`

Deseas saber cuántos estudiantes obtuvieron la nota 7. Usa count() para mostrarlo por pantalla.

Muestra por pantalla la nota más alta sin usar la función max(), sino que primero ordenas la lista (cópiala si no quieres perder el orden original) y luego tomas el último elemento con slicing.

Crea una list comprehension que genere una nueva lista notas_aprobadas con las notas mayores o iguales a 5. Muestra dicha lista.

(Opcional) Ordena la lista original con sort() en orden descendente y muestra el resultado.

````python
notas = [7, 9, 10, 6, 7, 8, 9]
print(notas.count(7))  # Usar paréntesis para llamar al método count
notas.sort()           # Ordenar la lista en orden ascendente
print(notas[-1])       # Mostrar la nota más alta (último elemento tras ordenar)
notas_aprobadas = [nota for nota in notas if nota >= 5]
print(notas_aprobadas)  # Mostrar solo las notas aprobadas (>=5)
````

````python
notas = [7, 9, 5, 10, 6, 7, 8, 9]

# 1. Contar cuántos sacaron un 7
cantidad_sietes = notas.count(7)
print(f"Estudiantes con nota 7: {cantidad_sietes}")

# 2. Nota más alta ordenando (sin usar max)
# Hacemos una copia para no alterar el orden original de la lista 'notas' todavía
notas_ordenadas = notas.copy() 
notas_ordenadas.sort() 
nota_maxima = notas_ordenadas[-1] # El último elemento es el mayor
print("Nota más alta:", nota_maxima)

# 3. List comprehension para aprobados (>= 5)
notas_aprobadas = [nota for nota in notas if nota >= 5]
print("Lista de aprobados:", notas_aprobadas)

# 4. (Opcional) Ordenar descendente
notas.sort(reverse=True)
print("Notas ordenadas descendente:", notas)
````

````python
tareas = ["limpiar", "estudiar", "comprar", "cocinar"]
tareas.insert(1, "pagar facturas")  # Inserta en la posición 1
tarea_realizada = tareas.pop(-1)  # Elimina y devuelve la última tarea
print(f"Aacabo de terminar la tarea: {tarea_realizada}")
print(tareas[1:3])  # Mostrar las tareas en las posiciones 1 y 2 (índices 1 y 2)
tareas_mayusculas = [tarea.upper() for tarea in tareas]  # Convertir todas las tareas a mayúsculas
print(tareas_mayusculas)
````

````python
tareas = ["limpiar", "estudiar", "hacer la compra"]

# 1. Insertar en posición 1
tareas.insert(1, "pagar facturas")
print("Tareas tras insertar:", tareas)

# 2. Extraer la última tarea con pop()
tarea_realizada = tareas.pop()
print(f"Acabo de finalizar la tarea: {tarea_realizada}")
print("Tareas restantes:", tareas)

# 3. Slicing de las dos primeras
primeras_dos = tareas[:2]
print("Las dos primeras tareas son:", primeras_dos)

# 4. List comprehension a mayúsculas
tareas_mayusculas = [t.upper() for t in tareas]
print("Tareas en mayúsculas:", tareas_mayusculas)
````

## 4. Bonus Track: Menú interactivo con while y condicionales
Crea un pequeño programa que gestione una lista de contactos con nombres (o puede ser otro tipo de datos). Debe funcionar de la siguiente forma:

Empieza con una lista vacía contactos = [].

Muestra un menú en pantalla con opciones:

markdown
Copiar
Editar
1. Añadir contacto
2. Mostrar contactos
3. Buscar contacto
4. Eliminar contacto
5. Salir
Usa un bucle while para que, hasta que el usuario elija la opción 5 (Salir), se repitan las siguientes acciones:

Opción 1: Pide al usuario que introduzca un nombre y añádelo a la lista con append().

Opción 2: Muestra la lista de contactos. (Si está vacía, indica “No hay contactos”.)

Opción 3: Pide un nombre a buscar en la lista. Si existe, indica su índice (usando index()). Si no existe, indica “No encontrado”.

Opción 4: Pide un nombre a eliminar. Si existe en la lista, elimínalo con remove(). Si no existe, indica “Ese contacto no está en la lista”.

Opción 5: Muestra “Saliendo del programa” y rompe el bucle.

Asegúrate de gestionar posibles errores (por ejemplo, si el usuario pide eliminar un contacto que no existe).

Valida el input del usuario para que si introduce una opción no válida, se le avise y se muestre el menú otra vez.

Pista: Recuerda que input() siempre devuelve strings. Para convertir a número, usa int(input()). Para comparaciones con nombres, trabajarás con strings.

````python
contactos = []

while True: # Bucle infinito hasta que el usuario decida salir
    print("\n--- GESTIÓN DE CONTACTOS ---")
    print("1. Añadir contacto")
    print("2. Mostrar contactos")
    print("3. Buscar contacto")
    print("4. Eliminar contacto")
    print("5. Salir")
    
    opcion = input("Elige una opción: ")

    if opcion == "1":
        nombre = input("Introduce el nombre del contacto: ")
        # Podríamos convertir a capitales para evitar duplicados por mayúsculas
        contactos.append(nombre) 
        print(f"Contacto '{nombre}' añadido.")

    elif opcion == "2":
        print("\n--- LISTA DE CONTACTOS ---")
        if len(contactos) == 0:
            print("No hay contactos registrados.")
        else:
            for contacto in contactos: #esto dice: para cada contacto en la lista de contactos, imprimelo
                print(f"- {contacto}")

    elif opcion == "3":
        buscar = input("¿A quién buscas?: ")
        if buscar in contactos:
            indice = contactos.index(buscar) # si el contacto existe, obtenemos su índice,index es para saber la posicion
            print(f"Contacto encontrado en la posición (índice): {indice}")
        else:
            print("No encontrado.")

    elif opcion == "4":
        eliminar = input("Nombre a eliminar: ")
        if eliminar in contactos: #esto dice: si el contacto esta en la lista eliminalo
            contactos.remove(eliminar)
            print("Contacto eliminado correctamente.")
        else:
            print("Ese contacto no está en la lista.")

    elif opcion == "5":
        print("Saliendo del programa...")
        break

    else:
        print("Opción no válida, por favor elige entre 1 y 5.")
````


3. **Enumerar**  
   - Tienes la lista `nombres = ["Ana", "Luis", "Marta", "Pedro"]`.  
   - Usa `enumerate()` para **iterar** sobre esta lista e imprimir por pantalla:  
     ```
     Persona 1: Ana
     Persona 2: Luis
     ...
     ```
   - Recuerda que puedes poner `start=1` si quieres enumerar desde 1.



````python
nombres = ["Ana", "Luis", "Marta", "Pedro"]
for i, nombre in enumerate(nombres, start=1):#esto quiere decir: para cada indice y nombre en la lista de nombres, imprimelo
    print(f"Índice {i}: {nombre}")
    
````

````python
nombres = ["Ana", "Luis", "Marta", "Pedro"]

# Usamos start=1 para que empiece a contar desde 1
for i, nombre in enumerate(nombres, start=1):
    print(f"Persona {i}: {nombre}")

# Salida esperada:
# Persona 1: Ana
# Persona 2: Luis
# ...
````

4. **Zip**  
   - Tienes estas dos listas:  
     ```python
     marcas = ["Ford", "Volkswagen", "Audi", "Seat"]
     paises = ["EEUU", "Alemania", "Alemania", "España"]
     ```
   - Usando `zip()`, imprime en pantalla tuplas del tipo:  
     ```
     ("Ford", "EEUU")
     ("Volkswagen", "Alemania")
     ...
     ```
   - Después, convierte el resultado de `zip(marcas, paises)` en una **lista de tuplas** (usa `list()`).

````python
marcas = ["Ford", "Toyota", "BMW", "Audi"   ]
paises = ["EEUU", "Alemania", "Japón", "Italia"]
for marca,pais in zip(marcas,paises):
    print(f"La marca {marca} es de {pais}")
listilla_tuplas =(list(zip(marcas,paises))) 
priny(f"Lista de tuplas marca-país: {listilla_tuplas}")


    
````

````python
marcas = ["Ford", "Volkswagen", "Audi", "Seat"]
paises = ["EEUU", "Alemania", "Alemania", "España"]

# 1. Imprimir tuplas usando un bucle
print("--- Bucle For ---")
for coche in zip(marcas, paises):
    print(coche) 
    # Imprime: ('Ford', 'EEUU'), etc.

# 2. Convertir a lista de tuplas
# Nota: Si intentas hacer list() sobre el mismo zip que ya recorriste arriba, saldrá vacío.
# Es mejor crear el zip de nuevo dentro del list() o hacerlo antes.
lista_tuplas = list(zip(marcas, paises))

print("\n--- Lista de Tuplas ---")
print(lista_tuplas)
# Salida: [('Ford', 'EEUU'), ('Volkswagen', 'Alemania'), ...]
````
