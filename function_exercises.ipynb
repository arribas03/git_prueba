## Exercise 1: Days of the week
Create a function that receives a number (1-7) and returns the name of the day (1 = Monday). If invalid, return "Invalid number".

**Signature:** `def get_day_name(day_number: int) -> str:`

````python

def get_day_name(day_number: int) -> str:
    days = {
        1: "Monday",
        2: "Tuesday",
        3: "Wednesday",
        4: "Thursday",
        5: "Friday",
        6: "Saturday",
        7: "Sunday"
    }
    return days.get(day_number, "Invalid day number") #get es para obtener el valor asociado a la clave day_number, si no existe devuelve "Invalid day number"

#con get nos ahorramos tener que hacer un if else para comprobar si la clave existe

get_day_name(8)  # Returns "Invalid day number"

````

## Exercise 2: Inverted Pyramid
Create a function that takes an integer `n` and prints an inverted pyramid of numbers.

**Signature:** `def print_pyramid(n: int) -> None:`

Example for n=4:
```
1 2 3 4
1 2 3
1 2
1
```

````python
def print_pyramid(n: int) -> None:
    for i in range(n, 0, -1): #esto dice que i va desde n hasta 1, decrementandose de 1 en 1
        for j in range(1, i + 1):
            print(j, end=" ") # end=" " hace que no se haga un salto de linea despues de imprimir
        print()  # Salto de lÃ­nea despuÃ©s de cada fila
print_pyramid(5)
````

````python
def print_pyramid(n: int) -> None:
    for i in range(n, 0, -1): #por cada fila i  imprime numeros del 1 al i
        # Generamos nÃºmeros del 1 al i
        for j in range(1, i + 1): #por cada fila j imprime numeros del 1 al i
            print(j, end=" ") #end=" " hace que no se haga un salto de linea despues de imprimir
        print()  # Salto de lÃ­nea al terminar cada fila

print_pyramid(5)
````

````python
def invert_pyramid(n: int) -> None:
    for index in range(n):
        print(*list(range(n+1-index))[1:])

invert_pyramid(5)
````

## Exercise 3: Number Comparator

**Signature:** `def compare_numbers(a: float, b: float) -> str:`

````python
def compare_numbers(a: float, b: float) -> str:
    if a > b:
        return f"{a} is larger than {b}"
    elif a < b:
        return f"{a} is less than {b}"
    elif a == b:
        return f"{a} is equal to {b}"
    else:
        return "Invalid input" #por ejemplo si se pasan strings en vez de numeros
    
input_n1 = float(input("Enter the first number: "))
input_n2 = float(input("Enter the second number: "))
compare_numbers(input_n1, input_n2)



````

## Exercise 4: Letter Counter
Create a function that receives a text and a character. Return how many times that character appears (case-insensitive).

**Signature:** `def count_letter(text: str, char: str) -> int:`

````python
def count_letter(text: str, char: str) -> int: #esto dice que la funcion recibe un string y un caracter y devuelve un entero
    count = 0
    for letter in text: #esto dice que por cada letra en el texto
        if letter == char: #esto dice que si la letra actual es igual al caracter buscado
            count += 1 #incrementa el contador en 1, es igual a count = count + 1
    return count

input_1 = input("Enter a text: ").lower()
input_2 = input("Enter a character to count: ").lower()
result = count_letter(input_1, input_2)
print(f"The character '{input_2}' appears {result} times in the text '{input_1}'")
            
        
    






````

````python
def count_letter(text: str, char: str) -> int:
    # Convertimos ambos a minÃºsculas para que sea case-insensitive,esto significa que no importa si la letra es mayuscula o minuscula
    return text.lower().count(char.lower())

input_text = input("Escribe una frase")
input_letter = input("Escribe una letra")
count_letter(input_text, input_letter)
````

## Exercise 5: Letter Frequency
Write a function that returns a **dictionary** with the count of each letter in a string.

**Signature:** `def letter_frequency(text: str) -> Dict[str, int]:`

Example: `"hello"` -> `{'h': 1, 'e': 1, 'l': 2, 'o': 1}`

````python
def letter_frequency(text: str) -> dict[str, int]:
    frequency = {}
    for letter in text:
        if letter in frequency:
            frequency[letter] += 1
        else:
            frequency[letter] = 1
    return frequency
input_text = input("Enter a text: ").replace(" ", "").lower() #remove spaces and convert to lowercase
frequency = letter_frequency(input_text)
print(f"Letter frequency for '{input_text}': {frequency}")



````

````python
from typing import Dict
def letter_frequency(text: str) -> Dict[str, int]:
    frecuencia = {}
    for letra in text:
        if letra in frecuencia:
            frecuencia[letra] += 1
        else:
            frecuencia[letra] = 1
    return frecuencia
phrase = input("Enter a phrase")
letter_frequency(phrase)
````

## Exercise 6: List Management
Write a function to modify a list. It receives:
1. The list.
2. A command ("add" or "remove").
3. The element (optional, default `None`).

**Signature:** `def manage_list(my_list: list, command: str, element: Any = None) -> list:`

````python

from typing import Any

def manage_list(my_list: list, command:str, element: Any = None) -> list: #any significa que puede ser de cualquier tipo
    if command == "add":
        my_list.append(element)
    elif command == "remove":
        if element in my_list:
            my_list.remove(element)
        else:
            print(f"Element {element} not found in the list.")
    return my_list

futbolistas = ["Messi", "Ronaldo", "Neymar"]
futbolistas = manage_list(futbolistas, "add", "Mbappe")
print(f"AÃ±adiendo uno nuevo: {futbolistas}")
futbolistas = manage_list(futbolistas, "remove", "Ronaldo")
print(f"Eliminando uno existente: {futbolistas}")
futbolistas = manage_list(futbolistas, "remove")
print(f"Intentando eliminar uno que no existe: {futbolistas}")
````

## Exercise 7: Join Words (`*args`)
Create a function that accepts an indefinite number of words and returns a sentence.

**Signature:** `def join_words(*words: str) -> str:`

````python
def join_words(*words: str) -> str:
    return ' '.join(words) #esto dice que une las palabras con un espacio entre ellas
join_words("Data", "Science", "is", "fun!")

lambda_join = lambda *words: ' '.join(words)
lambda_join("Lambda", "functions", "are", "cool!")
````

````python
def join_words(*words: str) -> str:
    # Usamos join para unir todas las tuplas de palabras con un espacio
    return " ".join(words)

join_words("Hola", "que", "tal")
````

## Exercise 8: Recursive Fibonacci
Create a **recursive** function to calculate the n-th Fibonacci number.

**Signature:** `def fibonacci(n: int) -> int:`

````python
def fibonacci(n: int) -> int:
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
fibonacci(5)  # deveulve 21 porque seria la secuencia 0, 1, 1, 2, 3, 5, 8, 13, 21,  

#la formula es F(n) = F(n-1) + F(n-2), por ejemplo F(8) = F(7) + F(6) = 13 + 8 = 21, por ejemplo F(5) = F(4) + F(3) = 5 + 3 = 8
#las funciones recursivas son funciones que se llaman a si mismas, en este caso fibonacci se llama a si misma para calcular los valores anteriores de la secuencia




````

````python
def fibonacci(n: int) -> int:
    #(0, 1, 1, 2, 3, 5, 8, 13...) 
    if n <= 1: #Para el 0 y el 1
        return n
    else:
        return fibonacci(n-1)+fibonacci(n-2)
      
fibonacci(3)

````

## Exercise 9: Geometry
Define functions to calculate area for: square, triangle, and circle.

**Signatures:**
* `def square_area(side: float) -> float:`
* `def triangle_area(base: float, height: float) -> float:`
* `def circle_area(radius: float) -> float:`

````python
import math #si no se importa math, no podemos usar math.pi

def square_area(side: float) -> float:
    return side ** 2

def triangle_area(base: float, height: float) -> float:
    return (base * height) / 2

def circle_area(radius: float) -> float:
    return math.pi * (radius ** 2) #esto es el area del circulo: pi por radio al cuadrado
print(square_area(4))    # 16
print(triangle_area(4, 5))  # 10
print(round(circle_area(3), 2))    # 28.27, para redondear a 2 decimales
````
