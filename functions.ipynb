## 3. The `return` statement

Las funciones devuelven valores. Si no hay `return`, devuelven `None`.

````python
def km_to_miles(km_distance: float) -> float:
    miles = km_distance * 0.621371
    return miles

# Storing the result in a variable
distance = km_to_miles(10.0)
print(f"10 km are {distance} miles")
````

### 3.1 Multiple Return
Python devuelve m煤ltiples valores como una tupla.

````python
def calculate_square_and_cube(number: int) ->  tuple[int, int]:
    square = number ** 2
    cube = number ** 3
    return square, cube

res_sq, res_cb = calculate_square_and_cube(3) #res_sq recibe el valor del cuadrado y res_cb el del cubo
print(f"Square: {res_sq}, Cube: {res_cb}")
````

## 4. Argument Types

### 4.1 Positional and Keywords
Podemos pasar argumentos por orden o por nombre.

````python
def describe_person(name: str, age: int, city: str) -> None:
    print(f"{name} is {age} years old and lives in {city}.")

# Positional (Order matters)
describe_person("Lucy", 30, "Madrid")

# Keywords (Order doesn't matter)
describe_person(city="Barcelona", name="Charles", age=25) #si lo desordenas no pasa nada porque usas keywords
````

### 4.2 Default Values
Valores por defecto si el usuario no env铆a nada.

````python
def make_coffee(sugar: bool = False) -> None:
    if sugar:
        print("Coffee with sugar ready ")
    else:
        print("Black coffee ready ")

make_coffee()        # Uses default (False)
make_coffee(True)    # Overwrites default, es decir le pasa True en vez de False
````

````python
def testing_words(words: str = "Example Text") -> None:
    print(words)

testing_words("Una frase de ejemplo")
````

## 5. Variable Arguments (`*args` & `**kwargs`)

* **`*args`**: Tupla de argumentos posicionales.
* **`**kwargs`**: Diccionario de argumentos nombrados.

````python
# Example using *args (Tuple)
def sum_all(*numbers: int) -> int:#el asterisco indica que numbers es una tupla de enteros,le podemos pasar varios elementos separados por comas
    print(f"Received: {numbers} (Type: {type(numbers)})")
    return sum(numbers)

print("Total sum:", sum_all(1, 2, 3, 4, 10))
````

````python
# Example using **kwargs (Dictionary)
def user_data(**data: Any) -> None: #los asteriscos dobles indican que data es un diccionario
    print(f"Data received: {data} (Type: {type(data)})")
    for key, value in data.items():
        print(f"- {key}: {value}")

user_data(name="John", age=33, vip=True)
````

## 6. Scope (Variable Visibility)

* **Local:** Dentro de la funci贸n.
* **Global:** Fuera de la funci贸n.

````python
x_str = "GLOBAL"  # Global variable

def test_function(x_str: str) -> None:
    x_str = "LOCAL"  # Local variable (shadows global)
    print("Inside:", x_str)


test_function(x_str)
print("Outside:", x_str)
#la primera imprime LOCAL y la segunda GLOBAL porque la variable local solo existe dentro de la funcion
````

## 7. Lambda Functions (Anonymous)

Funciones de una l铆nea sin nombre.

**Sintaxis:** `lambda args: expression`

````python
# Regular function
def double(x: int) -> int:
    return x * 2

# Equivalent Lambda
double_lambda = lambda x: x * 2

print(double(5))
print(double_lambda(5))
````

````python
# Capitalizar una lista de nombres
name_list = ["carlos martinez. elias navarro, julia mendes"]

def capitalize_list(list_name: list) -> list:
    return [s.title() for s in list_name] #la s es cada string en la lista

print(capitalize_list(name_list))

capilaze_lambda = lambda x_list : [s.title() for s in x_list]

print(capilaze_lambda(name_list))

````

## 9. Documentation (Docstrings)
Documentaci贸n est谩ndar explicando argumentos y retornos.

````python
def triangle_area(base: float, height: float) -> float:
    """
    Calculates the area of a triangle.
    
    Args:
        base (float): The base of the triangle.
        height (float): The height of the triangle.
        
    Returns:
        float: The calculated area.
    """
    return (base * height) / 2

# Checking help
help(triangle_area)
````
