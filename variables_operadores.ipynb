### Operadores Aritm√©ticos

- **Suma**: `+`
- **Resta**: `-`
- **Multiplicaci√≥n**: `*`
- **Divisi√≥n**: `/` (devuelve float)
- **Divisi√≥n entera**: `//` (descarta parte decimal)
- **M√≥dulo**: `%` (resto de la divisi√≥n)
- **Exponente**: `**`


```python
a = 12
b = 3

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicaci√≥n:", a * b, type(a * b))
print("Divisi√≥n:", a / b, type(a / b))
print("Divisi√≥n redondeada:", round(a / b), type(round((a / b))))
print("Divisi√≥n entera:", a // b)
print("M√≥dulo:", a % b)
print("Exponente", a ** b)
```

#### 3er error: Divisi√≥n por zero

```python
a = 10
b = 0
print(a / b)
```

### Operadores de Comparaci√≥n

- **Igual que**: `==`
- **Distinto de**: `!=`
- **Mayor que**: `>` / **Menor que**: `<`
- **Mayor o igual**: `>=` / **Menor o igual**: `<=`

```python
x = 3
y = 4
z = 2
print("¬øx es igual a y?", x == y)
print("¬øx es diferente a y?", x != y)
print("¬øx es mayor que y?", x > y)
print("¬øx es menor o igual a y?", x <= y)
print("¬øy est√° entre x y z", x < y < z)
```

### Operadores L√≥gicos

- `and`: verdadero si ambos operandos son verdaderos
- `or`: verdadero si al menos un operando es verdadero
- `not`: niega el valor de verdad


```python
a = True
b = False

print("a and b:", a and b)
print("a or b:", a or b)
print("not a:", not a)
```

Python es binario en esp√≠ritu (lo dejo como extra)

```python
# Comparaciones que muestran el comportamiento binario de los booleanos
print(f'False == 0?: {False == 0}')
print(f'True == 1?: {True == 1}')
print(f'True == 2?: {True == 2}')
print(f'False + True: {False + True}')
print(f'False - True: {False - True}')
print(f'True * 10: {True * 10}')
print(f'False * 10: {False * 10}')
```

## üéüÔ∏è Ejercicio Integrador 2: Entrada al cine y descuento (sin condicionales)

El precio de la entrada del cine es **14‚Ç¨**.

‚úÖ Las personas tienen un **descuento del 50%** si cumplen alguna de estas condiciones:

- **Menores de 12 a√±os**.
- **Mayores de 65 a√±os**.
- O si tienen una **tarjeta de cliente frecuente**.

---

### Datos:
- `age`: edad de la persona.
- `has_card`: `True` si tiene tarjeta de cliente frecuente, `False` si no.

---

### Actividades:

1. Calcula si **le corresponde el descuento** (muestra `True` o `False`).
2. Si quieres, calcula el precio final del boleto multiplicando el precio original por `0.5` si tiene descuento, o por `1` si no tiene descuento.  
   *(Sugerencia: puedes usar operadores aritm√©ticos y l√≥gicos, sin usar `if`).*

---

‚úÖ Ejemplo de salida esperada:

> ¬øTiene descuento? True Precio final de la entrada: 7.50

```python
age= int(input("Introduce tu edad: "))

print(age)

input_tarjeta = input("¬øEres cliente? (si/no): ")

tarjeta_cliente = (input_tarjeta == "si")


descuento = age <= 12 or age >= 65 or tarjeta_cliente == True

precio = 14

print("¬øTienes descuento?", descuento)

print("Precio de la entrada: ", precio *(1-(0.5*descuento)))
```

### üÜï M√©todos de String (Data Cleaning)
En Data Science, los datos nunca vienen limpios. Aqu√≠ unos trucos b√°sicos para normalizar texto.

```python
correo = "   USUARIO@GMAIL.COM   "

print(f"Original: '{correo}'")
print(f"Sin espacios extra: '{correo.strip()}'")
print(f"Todo min√∫sculas: '{correo.strip().lower()}'")

name = "paula rodr√≠guez sim√≥n"
print(f"Original: {name}")
print(f"May√∫sculas {name.upper()}")
print(f"Capitalizado primera palabra: {name.capitalize()}")
print(f"Capitalizado como t√≠tulo: {name.title()}")
```

### Tips importantes

- Las variables deben nombrarse con cuidado y evitar palabras reservadas (ej. `def`, `class`, `if`, `for`, etc.).
- Se recomienda usar nombres descriptivos para mejorar la legibilidad del c√≥digo.

### Reto 1: Calculadora de IMC
Pide por `input` el peso y la altura de un alumno y muestra su IMC con 2 decimales.

```python
height= float(input("Indica tu altura en cm: "))

weight= float(input("Indica tu peso en kg: "))

imc = weight / (height/100)**2

print(f"Tu IMC es: {imc:.2f}")
print(f"Tu IMC es: {round(imc,2)}")

```
